```python

import pandas as pd  # To work with arrays
import numpy as np   # To work with data frame
import matplotlib.pyplot as plt # Data Visualization 
%matplotlib inline
import seaborn as sns  # for using charts for data Visulaization

```

```python
from pandas.api.types import CategoricalDtype
```

```python
# Import data from file
df = pd.read_csv('Diwali Sales Data.csv',encoding ='unicode_escape')  

# encoding ='unicode_escape' to avoid error of thatb utf-8 can not read this file

df.shape   # to Get the no. of rows and coumns
```

```python
df.head()   #top 5 rows data
```

```python
# Data Cleaning
```

```python
df.info()
```

```python
# Deleting columns with 0 values using panda's drop

df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
# Checking For NULL Values
pd.isnull(df).sum()
```

```python
df.shape
```

```python
# Drop NULL Values :-> using dropna   
# inplace= True to perform action deirectly on data frame and save it
df.dropna(inplace= True)
```

```python
df.shape
```

```python
pd.isnull(df).sum()
```

```python
# Change data type :-> amount to whole number
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
# Description of data
df.describe()
```

```python
# Exploratory Data Analysis
```

```python
# Gender Count
```

```python
df.columns
```

```python
# ax = sns.countplot( x='Gender',data =df)  #countplot will count

# for bars in ax.containers:  # for showing the counted value above bars
#     ax.bar_label(bars)
# import seaborn as sns
# import matplotlib.pyplot as plt

# Assuming you have already loaded and prepared your DataFrame 'df'

# Create the countplot
ax = sns.countplot(x='Gender', data=df)

# Add labels above the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

# Optionally, you can set labels and customize the plot
ax.set_title('Count of Gender')
ax.set_xlabel('Gender')
ax.set_ylabel('Count')

# Show the plot
plt.show()

```

```python
# Ammount Vs Gender
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

```python
 # From above two graphs we can say that most of the buyers are female and even the purchasing pozwer of females are greater than man
```

```python
# Age vs Gender
# hue it will distingush and show the data on the basis of given parameter
```

```python
ax =sns.countplot(x ='Age Group',data=df,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total amount vs Age Group
```

```python
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
# From above Graphs we can say that most of the buyers are of age group 26-35
```

```python
# Total No. of Orders from top 10 states
```

```python
sales_states = df.groupby('State',as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,15)}) #set width and Height of graph
sns.barplot(x='State',y='Orders',data=sales_states)
```

```python
# Amount vs top 10 states 
```

```python
sales_states = df.groupby('State',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,15)}) #set width and Height of graph
sns.barplot(x='State',y='Amount',data=sales_states)
```

```python
# Marital_status
df.columns
```

```python
ax = sns.countplot(x='Marital_Status',data=df,hue='Gender')

sns.set(rc={'figure.figsize':(4,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Amount Vs Marital Status
sales_states = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(5,5)}) #set width and Height of graph
sns.barplot(x='Marital_Status',y='Amount',data=sales_states,hue='Gender')

```

```python
# From above graphs we can see that most of the buyers are Married(woman) and they have high purchasing power
```

```python
# Count vs Occupation
```

```python
ax =sns.countplot(x='Occupation', data =df)

sns.set(rc={'figure.figsize':(20,10)})
for bars in ax.containers:
        ax.bar_label(bars)
```

```python
# Amount vs Occupation
sales_oc = df.groupby(['Occupation','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Occupation',y='Amount',hue='Gender',data =sales_oc)
```

```python
# From the above graph we cam say that most of the buyers are from IT,Aviation and Healthcare Sector
```

```python
# Count vs Product Category
```

```python
ax= sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(25,10)})
```

```python
# Amount Vs Product Category
sales_p = df.groupby('Product_Category',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(35,15)})
sns.barplot(x='Product_Category',y='Amount',data=sales_p)
```

```python
# From the above graphs we can say that most of the sold products are from Food,Clothing and Electronics Category
```

```python
# Top 10 most Sold Products 
sales_pr = df.groupby('Product_ID',as_index =False)['Orders'].sum().sort_values(by='Orders',ascending =False).head(10)
sns.barplot(x='Product_ID',y='Orders',data=sales_pr)
sns.set(rc={'figure.figsize':(25,12)})
```

```python
# Same as above :-> another method
fig1,ax1= plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
